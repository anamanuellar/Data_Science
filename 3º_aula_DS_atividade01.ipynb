```python
df = pd.DataFrame( #1) Criar um DataFrame utilizando como fonte de dados um dicionário Python
    {
        "Matrícula": [1111, 2222, 3333, 4444, 5555],
        "Nome": ["João das Neves", "Joana Darc", "Leia", "Marcos", "Miguel A"],
        "Idade": [18, 19, 18, 28, 26],
        "Curso": ["Metaverso", "Dev Fullstack", "Dev Fullstack", "Fotografia", "Design"]
    }
)
```

```python
df.sample(5) #2) Imprimir todo o df com o método sample()
```

```python
df.loc[df.Idade == df.Idade.max()] #3) Qual(is) o(s) estudante(s) mais velho(s) ?
```

```python
df.loc[df.Idade == df.Idade.min()] #4) Qual(is) o(s) estudante(s) mais novo(s)?
```

```python
df.Idade.mean() #5) Qual a média de Idade dos estudantes?
```

```python
diferenca = df.Idade.max() - df.Idade.min() #6) Qual a diferença de idade entre o(s) estudante(s) mais velho(s) e o(s) mais novo(s) ?
diferenca
```
