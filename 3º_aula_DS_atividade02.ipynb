```python
import pandas as pd
```

```python
import numpy as np
```

```python
dfEmprestimo = pd.read_csv("previsao.csv") #1) Carregue para um DataFrame o Banco de Dados, 'previsao.csv' sobre previsão de Empréstimos.
```

```python
dfEmprestimo = dfEmprestimo.drop("Loan_ID", axis='columns') #3) Exclua a Coluna 'Loan_ID'
```

```python
colunas =["Genero", "Casado", "Dependentes", "Educacao", "Autonomo", "Renda", "Renda_Fiador", "Emprestimo", "Prazo", "Historico_Emprestimo", "Propriedade", "Status_Emprestimo"]
dfEmprestimo.columns = colunas #2) Renomeie as colunas do DataFrame
```

```python
dfEmprestimo.sample(10)
```

```python
dfEmprestimo.isnull().sum() #4) Verifique se há valores nulos nas colunas.
```

```python
#5) Substitua NaN nas seguintes colunas pelos respectivos valores:
dfEmprestimo['Genero'] = dfEmprestimo['Genero'].fillna('I') # Genero 'I'
```

```python
dfEmprestimo['Autonomo'] = dfEmprestimo['Autonomo'].fillna('No') #Autonomo 'No'
dfEmprestimo.Autonomo.sample(30)
```

```python
dfEmprestimo.Emprestimo.replace(np.nan, dfEmprestimo.Emprestimo.median()), #Emprestimo        Mediana ( median(Emprestimo) )
```

```python
dfEmprestimo['Prazo'] = dfEmprestimo['Prazo'].fillna(dfEmprestimo['Prazo'].mode())#Prazo Pagamento   Moda ( mode(Prazo Pagamento) )
```

```python
dfEmprestimo.Genero.replace("Male", "M", inplace= True), #Converta o que é "Male" em 'M' e o que é "female" em 'F'
dfEmprestimo.Genero.replace("Female", "F", inplace=True),
dfEmprestimo.sample(15)
```

```python
dfEmprestimo.Status_Emprestimo.replace("Y", "S", inplace= True), #Categoriza a Coluna 'Status Emprestimo' para 'S' ou 'N'
dfEmprestimo.sample(15)
```

```python
dfEmprestimo.Casado.replace("Yes", "S", inplace= True), #7) Categorize a Coluna 'Casado' para 'S' ou 'N'.
dfEmprestimo.Casado.replace("No", "N", inplace= True)
dfEmprestimo.sample(15)
```

```python
df = pd.DataFrame({'Emprestimo': [dfEmprestimo.Emprestimo.mean(), dfEmprestimo.Emprestimo.std(), dfEmprestimo.Emprestimo.median(), dfEmprestimo.Emprestimo.mode()[0]], 
                   'Prazo': [dfEmprestimo.Prazo.mean(), dfEmprestimo.Prazo.std(), dfEmprestimo.Prazo.median(), dfEmprestimo.Prazo.mode()[0]]}, index=['Mean', 'Std', 'Median', 'Mode'])
df
 #Crie um DataFrame Estatísticos com as funções: Média aritmética, desvio padrão, mediana e moda
    # das colunas 'Emprestimo' e 'Prazo de pagamento'
```
